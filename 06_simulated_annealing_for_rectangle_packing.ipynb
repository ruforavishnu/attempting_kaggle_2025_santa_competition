# Now that we have imported and ran all functions, classes and variables from our prev notebooks. lets start working on simulated annealing

```python
# This should already exist from previous notebooks
world_w = 100
world_h = 60

rects = [
    Rect(60, 10, id='A'),
    Rect(55, 12, id='B'),
    Rect(40, 30, id='C'),
    Rect(30, 35, id='D'),
    Rect(25, 25, id='E'),
    Rect(20, 20, id='F')
]

pack = pack_from_order(order_by_area(rects), world_w, world_h)
plot_rectangles(pack, world_w, world_h)
```

```python
import random
import math

def simulated_annealing(rects,    world_w,     world_h,     iterations=500,     T_start = 1000,     T_end = 1 ,    alpha = 0.995 ):
  current_order = order_by_area(rects)
  current_pack = pack_from_order(current_order, world_w, world_h)
  current_score = packing_score(current_pack)

  best_pack = current_pack
  best_score = current_score

  T = T_start

  for step in range(iterations):
    if T < T_end:
      break

    candidate_order = swap_two(current_order)
    candidate_pack = pack_from_order(candidate_order, world_w, world_h)


    if len(candidate_pack) < len(rects):
      T *= alpha
      continue

    candidate_score = packing_score(candidate_pack)
    delta = candidate_score - current_score

    if delta < 0:
      current_order = candidate_order
      current_pack = candidate_pack
      current_score = candidate_score

    else:
      acceptance_prob = math.exp(-delta / T)

      if random.random() < acceptance_prob:
        current_order = candidate_order
        current_pack = candidate_pack
        current_score = candidate_score



    if current_score < best_score:
      best_score = current_score
      best_pack = current_pack
      print(f'Step {step} : New best score = {best_score}')


    T *= alpha

  return best_pack, best_score




```

```python
best_pack , best_score  = simulated_annealing(rects, world_w, world_h, iterations=800, T_start = 1000, T_end = 1, alpha=0.995)

print('Best score found: ', best_score)
plot_rectangles(best_pack,  world_w, world_h)

```
