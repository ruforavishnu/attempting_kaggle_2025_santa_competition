```python
import random

def swap_two(rects):
  rects_new = rects[:]

  i, j = random.sample(   range(   len(rects)),    2)
  rects_new[i] , rects_new[j] = rects_new[j], rects_new[i]


  return rects_new


```

```python
import copy


def pack_from_order(rects, world_w,  world_h):
  rects_copy = copy.deepcopy(rects)

  return greedy_pack_with_rotation(rects_copy, world_w, world_h)

```

```python
def hill_climb(rects,  world_w,  world_h,   iterations=200):
  current_order = order_by_area(rects)
  current_pack = pack_from_order(current_order, world_w, world_h )

  current_score = packing_score(current_pack)


  for step in range(iterations):
    candidate_order = swap_two(current_order)
    candidate_pack = pack_from_order(candidate_order,  world_w, world_h)


    if len(candidate_pack) < len(rects):
      continue

    candidate_score = packing_score(candidate_pack)

    if candidate_score < current_score:
      current_order = candidate_order
      current_pack = candidate_pack
      current_score = candidate_score

      print(f'Step {step} : improved score -> {current_score}')



  return current_pack, current_score
```
