```python
import numpy as np
import matplotlib.pyplot as plt
from shapely.geometry import Polygon
from shapely.affinity import rotate

```

```python
def rotate_polygon(poly, angle_deg):
  return rotate(poly,  angle_deg,   origin='centroid', use_radians=False)

```

```python
points = [  (0,0),  (2,0),   (1,2)  ]

poly = Polygon(points)

rotated_30 = rotate_polygon(poly, 30)
rotated_90 = rotate_polygon(poly, 90)

fig, ax = plt.subplots(figsize=(5,5))


for p,c, lbl in [
    (poly, 'blue', 'Original'),
    (rotated_30, 'green', '30°'),
    (rotated_90, 'red', '90°')
]:
  x, y = p.exterior.xy
  ax.plot(x,y, label=lbl, color=c)


ax.set_aspect('equal')
ax.legend()
plt.show()



```

```python
def bounding_box(poly):
  minx, miny, maxx, maxy = poly.bounds
  return minx, miny, maxx, maxy
```

```python
for angle in [0, 30, 60, 90]:
  rp = rotate_polygon(poly, angle)
  print(f'Angle {angle}° bounds: ', bounding_box(rp))

```

```python
def collides(poly, others):
  for p in others:
    if poly.intersects(p):
      return True

  return False

```

```python
poly1 = Polygon([  (0,0),  (3,0),  (3,1), (0,1)])
poly2 = Polygon( [ (2,0),  (4,0),  (4,1)  , (2,1)  ])


rotated = rotate_polygon(poly2, 45)

fig,ax = plt.subplots(figsize=(5,5))

for p,c, lbl in [
    (poly1, 'blue', 'Poly A'),
    (rotated, 'red', 'Poly B rotated')
]:
  x,y = p.exterior.xy
  ax.plot(x,y,  label=lbl,  color=c)


ax.set_aspect('equal')
ax.legend()
plt.show()



print('Collision: ', poly1.intersects(rotated))

```

```python
def generate_rotations(poly, angles):
  return [rotate_polygon(poly, a) for a in angles]

```

```python
angles = list(range(0, 180, 30))
rotations = generate_rotations(poly, angles)

fig, ax = plt.subplots(figsize=(6,6))

for r in rotations:
  x,y = r.exterior.xy
  ax.plot(x,y, alpha=0.6)

ax.set_aspect('equal')
plt.title('Multiple rotations of polygons')
plt.show()

```

```python
def find_feasible_rotation(poly, placed, world_w, world_h, angles):
  for angle in angles:
    rp = rotate_polygon(poly, angle)
    minx, miny, maxx, maxy = rp.bounds

    if minx < 0 or miny < 0 or maxx > world_w or maxy > world_h:
      continue

    if not collides(rp, placed):
      return rp


  return None
```
